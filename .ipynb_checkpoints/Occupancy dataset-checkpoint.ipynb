```python
X_train=train.drop('Occupancy',axis=1)
y_train=train['Occupancy']

X_test1=test1.drop('Occupancy',axis=1)
y_test1=test1['Occupancy']


X_test2=test2.drop('Occupancy',axis=1)
y_test2=test2['Occupancy']

featurecombination=[('CO2', 'Light'),('CO2', 'Light','HumidityRatio'),('CO2', 'Light','Temperature'),
                    ('CO2', 'Light','HumidityRatio','Temperature')]
```

# logistic  regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
k_scores=[]
for f in featurecombination:
    print(f)
    X=X_train.loc[:,f]
    X_t1=X_test1.loc[:,f]
    X_t2=X_test2.loc[:,f]
    clf=LogisticRegression()
    clf.fit(X,y_train)
    print ("train")
    print(classification_report(y_train,clf.predict(X)))
    print(confusion_matrix(y_train,clf.predict(X)))
    print ("test1")
    print(classification_report(y_test1,clf.predict(X_t1)))
    print(confusion_matrix(y_test1,clf.predict(X_t1)))
    print ("test2")
    print(classification_report(y_test2,clf.predict(X_t2)))
    print(confusion_matrix(y_test2,clf.predict(X_t2)))
    print("\n\n")
```

```python
clf.fit(X_train.loc[:,('Light','CO2')],y_train)
clf.score(X_train.loc[:,('Light','CO2')],y_train)
```
